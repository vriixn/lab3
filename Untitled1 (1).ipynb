```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("student_performance.csv")
df.head()
```

Check Data Structure

```python
df.info()
df.shape
df.describe()
```

this dataset contains 6 columns

Missing Values

```python
df.isnull().sum()
```

No missing values found.

Remove Duplicates

```python
df.duplicated().sum()
```

No duplicate rows were found in the dataset.

Visualizations

```python
df.hist(figsize=(10,8))
plt.show()
```

```python
sns.scatterplot(x='weekly_self_study_hours', y='total_score', data=df)
plt.show()
```

Students who study more hours tend to achieve higher total scores.

```python
sns.scatterplot(x='attendance_percentage', y='total_score', data=df)
plt.show()
```

The data points are widely scattered, indicating no strong linear relationship between the two variables.
Although higher attendance sometimes corresponds to higher scores, the trend is not clear or consistent.
Overall, attendance alone does not strongly predict students total scores

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.show()
```

There is a strong positive correlation (0.81) between weekly study hours and total score, indicating that students who study more tend to achieve higher scores.

```python
df.groupby('grade')['total_score'].mean()
```

Students with grade A have the highest average total score.
